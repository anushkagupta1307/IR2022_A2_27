```python
#import necessary libraries
import pydrive
import re
from nltk.corpus import stopwords
from gensim.parsing.preprocessing import remove_stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
ps = PorterStemmer()
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
import nltk
import numpy as np
import pandas as pd
#nltk library for preprocessing
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
stop_words = set(stopwords.words('english'))
```

```python
#This method takes the Data files as input, does preprocessing steps on the data and then creates a unigram inverted index
import os
def preprocessing_indexing():
  token_dict=dict()
  doc_docID_mapping=dict()

  directory = '/content/drive/MyDrive/Humor,Hist,Media,Food'

  docID=1
  for filename in os.scandir(directory):
      flag=0
      if filename.is_file():
          ccfile = open(filename.path, "r",encoding="utf8", errors="surrogateescape")
          doc_docID_mapping[docID]=filename.path

          for aline in ccfile:
              #Preprocessing steps included- 
              # Convert to lowercase
              aline = aline.lower()
              # Remove numerics
              aline = re.sub(r'\d+','',aline)
              # Remove whitespaces
              aline = re.sub(r'[^\w\s]','', aline)
              # Remove special characters
              aline= re.sub('[@_!#$%^&*()<>?/\|}{~:]','',aline)
              aline = aline.strip()
              #Remove stopwords
              aline = remove_stopwords(aline)
              #Convert to tokens
              words = word_tokenize(aline)
          
              #Create a dictionary for the index - the dictionary has the Doc ID as the key and list of words as the value
              for w in words:
                if docID not in token_dict:
                  list1=[ps.stem(w)]
                  token_dict[docID]=(list1)
                else:
                  token_dict[docID].append(ps.stem(w))

          docID=docID+1              


  return token_dict,doc_docID_mapping
```

```python
token_dict,doc_docID_mapping=preprocessing_indexing()
```

```python
#added all words to a list
vocab_list=[]
for key, value in token_dict.items():
  for i in range(len(value)):
    vocab_list.append(value[i])
    
```

```python
#getting unique words from the list
temp_set=set(vocab_list)
vocab_list_final=list(temp_set)
```

**Term Frequency Weighting Scheme**

```python
#Creating an empty matrix with docids as the rows and words as the columns
temp_test_zeros=np.zeros((len(token_dict),len(vocab_list_final)))
term_frequency=pd.DataFrame(data=temp_test_zeros,columns=vocab_list_final,dtype=float)
```

```python
#filling the matrix to populate values with term frequency
for i in range(len(term_frequency)):
  for j in range(len(vocab_list_final)):
    token_list=token_dict[i+1]
    number_of_rep=token_list.count(vocab_list_final[j])
    total_number_words=len(token_list)
    term_frequency.at[i,vocab_list_final[j]]=float(number_of_rep/total_number_words)

    
  
```

**Calculating IDF**

```python
#Create an empty dictionary- count_inverse to get the number of times a word exists- count of words
count_inverse=dict()
for i in range(len(vocab_list_final)):
  for j in range(len(token_dict)):
      if vocab_list_final[i] not in count_inverse:
          temp_list=token_dict[j+1]
          exist_count=temp_list.count(vocab_list_final[i])
          # checking if it is more then 0
          if exist_count > 0:
            count_inverse[vocab_list_final[i]]=1
              
      else:
        temp_list=token_dict[j+1]
        exist_count=temp_list.count(vocab_list_final[i])
        # checking if it is more then 0
        if exist_count > 0:
          count_inverse[vocab_list_final[i]]=count_inverse[vocab_list_final[i]]+1

```

```python
#Calculating IDF
inverted_count_inverse=dict()
for key, value in count_inverse.items():
  temp=float(float(len(doc_docID_mapping))/float(count_inverse[key]))
  inverted_count_inverse[key]=temp
```

```python
#Calculating IDF
import math
log_count_inverse=dict()
for key, value in inverted_count_inverse.items():
  temp=math.log2(inverted_count_inverse[key])
  log_count_inverse[key]=temp
```

```python
#Creating empty matrix for storing IDF value of each word
temp_test_zeros=np.zeros((1,len(vocab_list_final)))
inverse_document_frequency=pd.DataFrame(data=temp_test_zeros,columns=vocab_list_final,dtype=float)
```

```python
#Populate the matrix- with columns as words and a single row for IDF
for i in range(len(vocab_list_final)):
  inverse_document_frequency.at[0,vocab_list_final[i]]=log_count_inverse[vocab_list_final[i]]
```

```python
#This method takes the Data files as input, does preprocessing steps on the data and then creates a unigram inverted index
import os
def preprocessing_indexing_for_binary():
  binary_term_dict=dict()
  

  directory = '/content/drive/MyDrive/Humor,Hist,Media,Food'

  docID=1
  for filename in os.scandir(directory):
      flag=0
      if filename.is_file():
          ccfile = open(filename.path, "r",encoding="utf8", errors="surrogateescape")
          doc_docID_mapping[docID]=filename.path

          for aline in ccfile:
              #Preprocessing steps included- 
              # Convert to lowercase
              aline = aline.lower()
              # Remove numerics
              aline = re.sub(r'\d+','',aline)
              # Remove whitespaces
              aline = re.sub(r'[^\w\s]','', aline)
              # Remove special characters
              aline= re.sub('[@_!#$%^&*()<>?/\|}{~:]','',aline)
              aline = aline.strip()
              #Remove stopwords
              aline = remove_stopwords(aline)
              #Convert to tokens
              words = word_tokenize(aline)
          
              #Create a dictionary for the index - the dictionary has the Root word(Stemming) as key and list of docIds as the value
              for w in words:
                if ps.stem(w) not in binary_term_dict:
                  list1=[docID]
                  binary_term_dict[ps.stem(w)]=(list1)
                else:
                  binary_term_dict[ps.stem(w)].append(docID)

          docID=docID+1              


  return binary_term_dict
```
